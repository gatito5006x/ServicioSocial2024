<center><h3><font face="Times New Roman">INEGI</h3>
<p style="text-align: justify;"><font face="Times New Roman">INEGI (Instituto Nacional de Estadstica, Geografa e Informtica) es una organizacin que se encarga de captar y difundir informacin de Mxico con relacin al territorio, ofreciendo datos de recursos, poblacin y economa que permite conocer caractersticas del territorio mexicano importantes en la toma de decisiones.</p>
<p style="text-align: justify;"><font face="Times New Roman">INEGI es una de las fuentes gubernamentales de suma importancia a la hora de consultar datos geoespaciales vectoriales o rster. En donde podemos disponer de una variedad de satlites con gran capacidad y precisin, de los que se puede mencionar: GeoEye-1, Landsat-8, RapidEye, Sentinel-2A, entre otros.</p>
<p style="text-align: justify;"><font face="Times New Roman">A travs de la herramienta Sistema de Consulta el usuario puede buscar el rea de inters que desea descargar, adems de poder aplicar filtros o seleccionar el satlite que se desea trabajar. Se pueden descargar achivos TIFF, SHP, GIF, entre otros.</p>
<p style="text-align: justify;"><font face="Times New Roman">En el siguiente cdigo de ejemplo se emplea geopandas para cargar y visualizar un archivo shapefile que contiene Informacin vectorial de localidades amanzanadas y nmeros exteriores 2023. Texcoco. Lo que facilita el anlisis y visualizacin de los datos geoespaciales proporcionados por INEGI.</p>

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Ruta al archivo shapefile descargado de INEGI
shapefile_path = '150990001a.shp'

# Leer el archivo shapefile
gdf = gpd.read_file(shapefile_path)

# Mostrar las primeras filas del GeoDataFrame
print(gdf.head())

# Mostrar informacin del GeoDataFrame
print(gdf.info())

# Visualizar el mapa de los datos vectoriales
gdf.plot()
plt.title('Visualizacin del Shapefile de INEGI')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.show()
```

<center><h3><font face="Times New Roman">Landsat</h3>
<p style="text-align: justify;"><font face="Times New Roman">Landsat es un programa conjunto de la NASA y el Servicio Geolgico de los Estados Unidos (USGS) que proporciona datos satelitales de la superficie terrestre desde 1972. Los satlites Landsat son una herramienta crucial en el campo del geoprocesamiento, ya que proporcionan imgenes de alta resolucin y una larga serie temporal, lo que permite el monitoreo y anlisis de cambios en la superficie terrestre a lo largo del tiempo.</p>
<font face="Times New Roman"><p style="text-align: justify;">A lo largo de su trayectoria Landsat ha tenido distintas versiones, actualmente la ltima versin es Landsat 9, sin embargo, cada una de ellas proporcionan imgenes con una resolucin espacial de hasta 30 metros, adecuada para el monitoreo detallado de cambios en la superficie terrestre.</p>
<p style="text-align: justify;"><font face="Times New Roman">Adems, los sensores de Landsat capturan datos en mltiples bandas espectrales, incluyendo bandas visibles, infrarrojas y trmicas, lo que permite una amplia gama de aplicaciones, desde la agricultura y la silvicultura hasta la gestin de recursos hdricos y el monitoreo ambiental.</p>
<font face="Times New Roman"><p style="text-align: justify;">Los datos de Landsat estn disponibles de forma gratuita, lo que democratiza el acceso a informacin crucial para la investigacin y la gestin de recursos. A travs de distintas plataformas se pueden acceder a ciertos satlites Landsat: USGS Earth Explorer, NASA Earthdata, Google Earth Engine, INEGI, entre otros.</p>


<center><h3><font face="Times New Roman">Sentinel</h3></center>
<font face="Times New Roman"><p style="text-align: justify;">Es una serie de misiones satelitales desarrolladas por la Agencia Espacial Europea (ESA) como parte del programa Copernicus. Estas misiones proporcionan datos de observacin de la Tierra para una amplia gama de aplicaciones, incluyendo la monitorizacin del medio ambiente, la gestin de desastres, y la investigacin climtica.</p>
<font face="Times New Roman"><p style="text-align: justify;">Cada misin Sentinel est diseada para cubrir diferentes aspectos de la observacin de la Tierra, utilizando diversos tipos de sensores:
<ul>
<li>Sentinel-1: Utiliza radar de apertura sinttica (SAR) para proporcionar imgenes en todas las condiciones meteorolgicas, tanto de da como de noche.</li>
<li>Sentinel-2: Utiliza un sensor ptico multiespectral de alta resolucin para capturar imgenes en 13 bandas espectrales.</li>
<li>Sentinel-3: Proporciona datos de ocano y tierra utilizando una combinacin de sensores pticos y de microondas.</li>
<li>Sentinel-4: Diseado para monitorear la composicin atmosfrica desde una rbita geoestacionaria.</li>
<li>Sentinel-5: Centrado en la observacin de la atmsfera desde una rbita baja terrestre.</li>
<li>Sentinel-6: Mide la altimetra del nivel del mar utilizando un altmetro de radar.</li>
</ul></p>
<font face="Times New Roman"><p style="text-align: justify;">Los datos de los Sentinel estn disponibles de manera gratuita en distintos portales, tales como: Google Earth Engine, Copernicus Data and Information Access Services (DIAS), Copernicus Open Access Hub, etc.</p>
