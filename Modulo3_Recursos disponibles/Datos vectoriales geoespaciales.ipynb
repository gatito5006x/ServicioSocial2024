<p style="text-align: justify;"><font face="Times New Roman">En la actualidad para disponer de recursos geoespaciales hay una variedad de fuentes en línea, existen datos que son gratuitos y de uso abierto para la comunidad, que generalmente son proporcionados por agencias gubernamentales, organizaciones internacionales o instituciones académicas. <center><h3><font face="Times New Roman">INEGI</h3>
<p style="text-align: justify;"><font face="Times New Roman">INEGI (Instituto Nacional de Estadística, Geografía e Informática) es una organización que se encarga de captar y difundir información de México con relación al territorio, ofreciendo datos de recursos, población y economía que permite conocer características del territorio mexicano importantes en la toma de decisiones.</p>
<p style="text-align: justify;"><font face="Times New Roman">INEGI es una de las fuentes gubernamentales de suma importancia a la hora de consultar datos geoespaciales vectoriales o ráster. En donde podemos disponer de una variedad de satélites con gran capacidad y precisión, de los que se puede mencionar: GeoEye-1, Landsat-8, RapidEye, Sentinel-2A, entre otros.</p>
<p style="text-align: justify;"><font face="Times New Roman">A través de la herramienta Sistema de Consulta el usuario puede buscar el área de interés que desea descargar, además de poder aplicar filtros o seleccionar el satélite que se desea trabajar. Se pueden descargar achivos TIFF, SHP, GIF, entre otros.</p>
<p style="text-align: justify;"><font face="Times New Roman">En el siguiente código de ejemplo se emplea geopandas para cargar y visualizar un archivo shapefile que contiene Información vectorial de localidades amanzanadas y números exteriores 2023. Texcoco. Lo que facilita el análisis y visualización de los datos geoespaciales proporcionados por INEGI.</p>

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Ruta al archivo shapefile descargado de INEGI
shapefile_path = '150990001a.shp'

# Leer el archivo shapefile
gdf = gpd.read_file(shapefile_path)

# Mostrar las primeras filas del GeoDataFrame
print(gdf.head())

# Mostrar información del GeoDataFrame
print(gdf.info())

# Visualizar el mapa de los datos vectoriales
gdf.plot()
plt.title('Visualización del Shapefile de INEGI')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.show()
```

<center><h3><font face="Times New Roman">Landsat</h3>
<p style="text-align: justify;"><font face="Times New Roman">Landsat es un programa conjunto de la NASA y el Servicio Geológico de los Estados Unidos (USGS) que proporciona datos satelitales de la superficie terrestre desde 1972. Los satélites Landsat son una herramienta crucial en el campo del geoprocesamiento, ya que proporcionan imágenes de alta resolución y una larga serie temporal, lo que permite el monitoreo y análisis de cambios en la superficie terrestre a lo largo del tiempo.</p>
<font face="Times New Roman"><p style="text-align: justify;">A lo largo de su trayectoria Landsat ha tenido distintas versiones, actualmente la última versión es Landsat 9, sin embargo, cada una de ellas proporcionan imágenes con una resolución espacial de hasta 30 metros, adecuada para el monitoreo detallado de cambios en la superficie terrestre.</p>
<p style="text-align: justify;"><font face="Times New Roman">Además, los sensores de Landsat capturan datos en múltiples bandas espectrales, incluyendo bandas visibles, infrarrojas y térmicas, lo que permite una amplia gama de aplicaciones, desde la agricultura y la silvicultura hasta la gestión de recursos hídricos y el monitoreo ambiental.</p>
<font face="Times New Roman"><p style="text-align: justify;">Los datos de Landsat están disponibles de forma gratuita, lo que democratiza el acceso a información crucial para la investigación y la gestión de recursos. A través de distintas plataformas se pueden acceder a ciertos satélites Landsat: USGS Earth Explorer, NASA Earthdata, Google Earth Engine, INEGI, entre otros.</p>


<center><h3><font face="Times New Roman">Sentinel</h3></center>
<font face="Times New Roman"><p style="text-align: justify;">Es una serie de misiones satelitales desarrolladas por la Agencia Espacial Europea (ESA) como parte del programa Copernicus. Estas misiones proporcionan datos de observación de la Tierra para una amplia gama de aplicaciones, incluyendo la monitorización del medio ambiente, la gestión de desastres, y la investigación climática.</p>
<font face="Times New Roman"><p style="text-align: justify;">Cada misión Sentinel está diseñada para cubrir diferentes aspectos de la observación de la Tierra, utilizando diversos tipos de sensores:
<ul>
<li>Sentinel-1: Utiliza radar de apertura sintética (SAR) para proporcionar imágenes en todas las condiciones meteorológicas, tanto de día como de noche.</li>
<li>Sentinel-2: Utiliza un sensor óptico multiespectral de alta resolución para capturar imágenes en 13 bandas espectrales.</li>
<li>Sentinel-3: Proporciona datos de océano y tierra utilizando una combinación de sensores ópticos y de microondas.</li>
<li>Sentinel-4: Diseñado para monitorear la composición atmosférica desde una órbita geoestacionaria.</li>
<li>Sentinel-5: Centrado en la observación de la atmósfera desde una órbita baja terrestre.</li>
<li>Sentinel-6: Mide la altimetría del nivel del mar utilizando un altímetro de radar.</li>
</ul></p>
<font face="Times New Roman"><p style="text-align: justify;">Los datos de los Sentinel están disponibles de manera gratuita en distintos portales, tales como: Google Earth Engine, Copernicus Data and Information Access Services (DIAS), Copernicus Open Access Hub, etc.</p>

        </ul></p>
