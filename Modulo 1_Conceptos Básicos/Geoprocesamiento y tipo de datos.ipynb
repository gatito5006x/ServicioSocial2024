<h2><font face="Times New Roman">Geoprocesamiento</h2>
<p style="text-align: justify;"><font face="Times New Roman">La necesidad mundial de manejar los datos de ubicación, delimitación, georreferencias, localización, entre otros, que tradicionalmente se hacía de una manera poco eficiente llevo al desarrollo de la Geomática, que hace referencia al conjunto de ciencias en las cuales de integran los medios para la captura, tratamiento, análisis, interpretación, difusión y almacenamiento de información geográfica. Es decir, la geomática se encarga del estudio de la superficie de la tierra a través de la informática (“GEO” Tierra y “MATICA” informática).</p>
<p style="text-align: justify;"><font face="Times New Roman">No es hasta 1969 que el termino es definido por Bernard Dubuisson, siendo todas las ciencias de base y a las tecnologías usadas para el conocimiento del territorio como la teledetección o percepción remota, Sistemas de Información Geográfica (SIG), Sistemas Globales de Navegación por Satélite (GNSS) y conocimientos relacionados.</p>
<p style="text-align: justify;"><font face="Times New Roman">De estas ciencias surge el termino de geoprocesamiento, un conjunto de técnicas y herramientas utilizadas para manipular, analizar y gestionar datos espaciales o geográficos. Estas técnicas permiten realizar operaciones complejas sobre información geoespacial, como la creación de mapas, el análisis de patrones espaciales, la modelización de fenómenos geográficos y la toma de decisiones basada en la ubicación.</p>
<p style="text-align: justify;"><font face="Times New Roman">El geoprocesamiento incluye una amplia gama de operaciones, entre las que se encuentran:
<ul>
<li>Análisis espacial: Evaluar la relación entre diferentes elementos geográficos, como la proximidad, la superposición y la conectividad.</li>
<li>Transformación de datos: Convertir datos de un formato o sistema de coordenadas a otro.</li>
<li>Consulta y selección: Extraer subconjuntos específicos de datos geográficos basados en criterios definidos.</li>
<li>Edición de datos: Modificar o actualizar información geoespacial existente.</li>
<li>Integración de datos: Combinar datos de múltiples fuentes para crear un conjunto de datos unificado.</li>
<li>Modelización y simulación: Predecir o simular fenómenos geográficos basados en datos existentes.</li>
    </ul></p>
<p style="text-align: justify;"><font face="Times New Roman">El uso de del geoprocesamiento es presente en diversas áreas como la planificación urbana, gestión de recursos naturales, investigación ambiental, la agricultura y muchos otros campos que requieran un análisis aéreo de la superficie.</p>

<p></p><font face="Times New Roman"><h2>Datos ráster</h2></font></p>
<p style="text-align: justify;"><font face="Times New Roman">Generalmente los datos continuos como la elevación, o la temperatura se almacenan en un dataset geográfico denominado ráster. Los datos espectrales, como las fotografías aéreas de los drones y las imágenes satelitales también se almacenan de esta manera. Este tipo de datos no asume que existen límites estrictos entre los objetos por contrario de la forma que lo hacen los conjuntos de datos vectoriales.</p>
<p style="text-align: justify;"><font face="Times New Roman">Consisten en una matriz de celdas (píxeles), donde cada celda tiene un valor que representa información geográfica, como la elevación, la temperatura o la cobertura del suelo, este modelo de SIG ráster se centra en las propiedades del espacio más que en la precisión de la localización.
Los datos ráster se componen de filas y columnas de celdas, cada celda almacena un valor único. Los datos ráster pueden ser imágenes (imágenes ráster), con un valor de color en cada celda (o píxel). Otros valores registrados para cada celda puede ser un valor discreto, como el uso del suelo, valores continuos, como temperaturas, o un valor nulo si no se dispone de datos.</p>
<p style="text-align: justify;"><font face="Times New Roman">El hecho de que el valor de los pixeles pueden variar continuamente permite que los ráster sean apropiados para datos que varían continuamente, como la elevación. En lugar de contar con geometrías individuales se cuenta con colección de pixelen que conforman una gran matriz bidimensional o tridimensional. Un dataset ráster se compone por bandas y cada una de estas bandas es una matriz bidimensional, el conjunto de bandas se convierte en una matriz 3D.</p>

<p style="text-align: justify;"><font face="Times New Roman">Para la manipulación de los datos rasterio es necesario hacer uso de bibliotecas que se disponen en la comunidad de Python. Rasterio es una biblioteca de Python que permite la lectura y escritura de datos ráster, así como la manipulación y el análisis de datos geoespaciales en formato ráster. Está construida sobre la biblioteca de software GDAL (Geospatial Data Abstraction Library), lo que le proporciona robustez y flexibilidad para trabajar con una amplia variedad de formatos de archivos ráster.</p>

<p></p><font face="Times New Roman"><h2>Datos vectoriales</h2></font></p>
<p style="text-align: justify;"><font face="Times New Roman">Si se desea trabajar con datos existentes, ya sea para resumir, editar, derivar nuevos datos o realizar algún tipo de analisis, se debe interactuar con los datos vectoriales. Los datos vectoriales son datos en donde las características geometricas se representan a tráves de puntos, líneas y polígonos.
<ul>
<li>Puntos: Representan ubicaciones específicas, como una ciudad, un árbol o una estación meteorológica.</li>
<li>Líneas: Representan elementos lineales, como carreteras, ríos o líneas eléctricas.</li>
<li>Polígonos: Representan áreas cerradas, como lagos, parcelas de tierra o zonas urbanas.</li>
</ul></p>
<p style="text-align: justify;"><font face="Times New Roman">Las entidades geográficas que tienen límites distintos, como las ciudades funcionan perfectamente como datos vectoriales, pero los datos continuos como la elevación, no. Se complicaría demasiado dibujar un solo poligono alrededor de todas la áreas con la misma elevación, en ese caso se implementaría el uso de datos ráster como se mencionó anteriormente. Sin embargo, muchos tipos de datos son excelentes candidatos para una representación vectorial, como las carreteras que se representan como líneas, los condados y estados son polígonos, y dependiendo de la escala del mapa, las ciudades se dibujan como puntos o polígonos.</p>
<p style="text-align: justify;"><font face="Times New Roman">En los datos vectoriales, el interés de las representaciones se centra en la precisión de localización de los elementos geográficos sobre el espacio y donde los fenómenos a representar son discretos, es decir, de límites definidos. Cada una de estas geometrías está vinculada a una fila en una base de datos que describe sus atributos.</p>
<p style="text-align: justify;"><font face="Times New Roman">Para el manejo de los datos vectoriales de las librerías más populares y versátiles se encuentra el Geopandas. 
Geopanas proporciona una extensión de las capacidades de pandas para permitir la manipulación y el análisis de datos espaciales de forma sencilla. Se basa en otras bibliotecas como Shapely, Fiona y Pyproj para proporcionar una amplia gama de funcionalidades.</p>

<p style="text-align: justify;"><font face="Times New Roman">Como se mencionó geopandas se basa en la estructura de datos Pandas, por lo que en el corazón de GeoPandas existen dos conceptos clave: Geoseries y GeoDataFrame.</p>
<p style="text-align: justify;"><font face="Times New Roman">GeoSeries es basicamente un vector en el que cada entrada de la serie es un conjunto de formas u objetos geométricos, estos pueden ser cualquier punto, lineas o formas compatibles con Shapely, una biblioteca de Python para la manipulación y análisis de objetos geométricos planos.
<p style="text-align: justify;"><font face="Times New Roman">A continuación un ejemplo sencillo para representar una Geoserie, en dónde se crea un conjunto un conjunto de Geoseries de tres puntos y se visualiza usando el método .plot().</p>


```python
from shapely.geometry import Point, Polygon
import geopandas as gpd

# Create a GeoSeries of points
points = gpd.GeoSeries([Point(1, 1), Point(2, 2), Point(3, 3)])

# Visualize the points
points.plot()
```

<p style="text-align: justify;"><font face="Times New Roman">Una GeoDataFrame es una estructura de datos tabular que contiene un conjunto de datos GeoSeries. Similiar a un DataFrame pandas, con la diferencia que cuenta con una columna extra en donde se almacenan los datos de las información geométrica. Esta columna adicional comunmente denominada "geometría", contiene los objetos geométricos y permite realizar las operaciones espaciales en el conjunto de datos.</p>
<p style="text-align: justify;"><font face="Times New Roman">En el siguiente código se muestra cómo crear una GeoDataFrame a partir de una lista de geometrías Shapely. Basicamente se combinan puntos y polígonos en un solo GeoDataFrame, despues graficarlo y visualizarlo.</p>

```python
# Define some points and a polygon
polygon = Polygon([(0, 0), (1, 1), (1, 0)])
points = [Point(0.5, 0.5), Point(1.5, 1.5)]

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame({'geometry': points + [polygon]})

# Plot the GeoDataFrame
gdf.plot(alpha=0.5, linewidth=2, edgecolor='k', color='cyan')
```
