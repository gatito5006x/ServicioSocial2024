<p style="text-align: justify;"><font face="Times New Roman">Como se mencionÃ³ geopandas se basa en la estructura de datos Pandas, por lo que en el corazÃ³n de GeoPandas existen dos conceptos clave: Geoseries y GeoDataFrame.</p>
<p style="text-align: justify;"><font face="Times New Roman">GeoSeries es basicamente un vector en el que cada entrada de la serie es un conjunto de formas u objetos geomÃ©tricos, estos pueden ser cualquier punto, lineas o formas compatibles con Shapely, una biblioteca de Python para la manipulaciÃ³n y anÃ¡lisis de objetos geomÃ©tricos planos.
<p style="text-align: justify;"><font face="Times New Roman">A continuaciÃ³n un ejemplo sencillo para representar una Geoserie, en dÃ³nde se crea un conjunto un conjunto de Geoseries de tres puntos y se visualiza usando el mÃ©todo .plot().</p>


```python
from shapely.geometry import Point, Polygon
import geopandas as gpd

# Create a GeoSeries of points
points = gpd.GeoSeries([Point(1, 1), Point(2, 2), Point(3, 3)])

# Visualize the points
points.plot()
```

<p style="text-align: justify;"><font face="Times New Roman">Una GeoDataFrame es una estructura de datos tabular que contiene un conjunto de datos GeoSeries. Similiar a un DataFrame pandas, con la diferencia que cuenta con una columna extra en donde se almacenan los datos de las informaciÃ³n geomÃ©trica. Esta columna adicional comunmente denominada "geometrÃ­a", contiene los objetos geomÃ©tricos y permite realizar las operaciones espaciales en el conjunto de datos.</p>
<p style="text-align: justify;"><font face="Times New Roman">En el siguiente cÃ³digo se muestra cÃ³mo crear una GeoDataFrame a partir de una lista de geometrÃ­as Shapely. Basicamente se combinan puntos y polÃ­gonos en un solo GeoDataFrame, despues graficarlo y visualizarlo.</p>

```python
# Define some points and a polygon
polygon = Polygon([(0, 0), (1, 1), (1, 0)])
points = [Point(0.5, 0.5), Point(1.5, 1.5)]

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame({'geometry': points + [polygon]})

# Plot the GeoDataFrame
gdf.plot(alpha=0.5, linewidth=2, edgecolor='k', color='cyan')
```
